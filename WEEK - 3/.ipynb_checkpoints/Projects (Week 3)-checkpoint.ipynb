```python
import cmath

def solve_cubic(a, b, c, d):
    # Calculate discriminant
    delta0 = b**2 - 3*a*c
    delta1 = 2*b**3 - 9*a*b*c + 27*a**2*d
    delta = delta1**2 - 4*delta0**3

    # Check number of roots and calculate them
    if delta > 0:  # One real root and two complex roots
        root1 = (-b + (delta1 + cmath.sqrt(delta))/(2*cmath.sqrt(delta0))) / (3*a)
        root2 = (-b - (delta1 + cmath.sqrt(delta))/(2*cmath.sqrt(delta0))) / (3*a)
        root3 = (-b - 2*(delta1 + cmath.sqrt(delta))/(3*cmath.sqrt(delta0))) / (3*a)
        return (root1, root2, root3)
    elif delta == 0:  # Three real roots (two are equal)
        if delta0 == 0:
            root1 = -b / (3*a)
            return (root1, root1, root1)
        else:
            root1 = (-b + cmath.sqrt(delta0)) / (3*a)
            root2 = root3 = -b / (3*a) - 2*root1
            return (root1, root2, root3)
    else:  # Three real roots (all distinct)
        if delta0 == 0:
            root1 = -b / (3*a)
            root2 = root3 = -2*root1
        else:
            w = (-1 + cmath.sqrt(3)*1j) / 2
            w2 = w**2
            root1 = (-b + (w*delta1 + w2*cmath.sqrt(delta))/(3*a*(w**2))) / (3*a)
            root2 = (-b + (w2*delta1 + w*cmath.sqrt(delta))/(3*a*(w**2))) / (3*a)
            root3 = (-b - (delta1 + w*cmath.sqrt(delta))/(3*a*w)) / (3*a)
        return (root1, root2, root3)

# Example usage
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))
d = int(input("Enter d: "))

roots = solve_cubic(a, b, c, d)
print("The roots are:", roots)

```

```python
def solve_quartic(a, b, c, d, e):
    # Calculate coefficients of resolvent cubic
    p = -b/(4*a)
    q = (b**2 - 3*a*c)/(8*(a**2))
    r = (-b**3 + 4*a*b*c - 8*(a**2)*d)/(256*(a**3))

    # Calculate discriminant of resolvent cubic
    delta0 = q**2 - 4*p*r
    if delta0 > 0:
        u1 = (-q + (delta0)**(1/2))/2
        u2 = (-q - (delta0)**(1/2))/2
    else:
        u1 = complex(-q/2, (abs(delta0))**(1/2)/2)
        u2 = complex(-q/2, -(abs(delta0))**(1/2)/2)

    # Calculate roots of quartic equation using Cardano's formula
    v1 = (u1**2 - 4*r/u1 - 2*p).real
    v2 = (u2**2 - 4*r/u2 - 2*p).real

    # Calculate discriminant of quartic equation
    delta1 = v1**2 - 4*(u1**2 - 4*r/u1)

    # Calculate roots of quartic equation
    if delta1 > 0:
        x1 = (-v1 + delta1**(1/2))/(2*u1)
        x2 = (-v1 - delta1**(1/2))/(2*u1)
        x3 = (-v2 + delta1**(1/2))/(2*u2)
        x4 = (-v2 - delta1**(1/2))/(2*u2)
        return (x1, x2, x3, x4)
    elif delta1 == 0:
        x1 = (-v1)/(2*u1)
        x2 = x3 = (-v2)/(2*u2)
        return (x1, x2, x3)
    else:
        x1 = (-v1 + cmath.sqrt(delta1))/(2*u1)
        x2 = (-v1 - cmath.sqrt(delta1))/(2*u1)
        x3 = (-v2 + cmath.sqrt(delta1))/(2*u2)
        x4 = (-v2 - cmath.sqrt(delta1))/(2*u2)
        return (x1, x2, x3, x4)

# Example usage
a = int(input("Enter a: "))
b = int(input("Enter a: "))
c = int(input("Enter a: "))
d = int(input("Enter a: "))
e = int(input("Enter a: "))

roots = solve_quartic(a, b, c, d, e)
print("The roots are:", roots)
```
